```python
import matplotlib.pyplot as plt

category_column = 'Category'
values_column = 'Values'

# Group data by the 'Category' and calculate the sum of 'Values' for each category
grouped_data = df.groupby('sex')['age'].sum()

# Plotting the bar chart
grouped_data.plot(kind='bar', color='skyblue')

# Adding labels and title
plt.xlabel(category_column)
plt.ylabel(values_column)
plt.title('Bar Chart of ' + values_column + ' by ' + category_column)
```

## Working | FY24 Sales vs GMB FY24 - Ratings

```python
import io

import pandas as pd
df = pd.read_excel(io.BytesIO(uploaded['DATA FY23 vs FY24 Sales and GMB.xlsx']))
```

```python
df.head()
```

```python
# @title FY24 Sales vs GMB FY24 - Ratings

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='FY24 Sales', y='GMB FY24 - Ratings', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
correlation = df['FY24 Sales'].corr(df['GMB FY24 - Ratings'])
correlation
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame

# Calculate the correlation matrix
corr_matrix = df[['Sales G%', 'GMB FY24 - Ratings G%', 'GMB FY24 - Review G%']].corr()

print(corr_matrix)
# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='viridis', fmt=".2f")

# Add title
plt.title('Correlation Heatmap')

# Show the plot
plt.show()
```
