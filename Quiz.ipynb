```python
def foo(x):
  try:
    return 10/x
  except ZeroDivisionError:
    return "Error"
  finally: # The finally block always executes, and it returns 0.
    return 0

result = foo(0)
print(result) # due to the "finally" block the o/p will always be zero.
```
